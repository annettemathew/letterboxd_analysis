```python
import pandas as pd
import string
from collections import Counter
import matplotlib.pyplot as plt

# Load data
REVIEWED_DATA_FILE = "Downloads/letterboxd-fishnet22-2024-08-06-11-04-utc/reviews.csv"
review_data = pd.read_csv(REVIEWED_DATA_FILE)

# Define a list of stop words
stop_words = set([
    'and', 'the','like', 'also'
    , 'im', 'rly', 'w', 'def', 'u', 'i', 'of', 'to'
    #, 'this', 'a', 'was', 'it', 'in', 'for', 'that'
    , 'she', 'think', '1', '2', '3', '4', 'a', 'this', 'was'
    , 'so'#, 'so', 'is', 'my', 'she', 'but', 'have', 'did', 
    #'be', 'one', 'at', 'all', 'on', 'as', 'not', 'with'
    #,'if', 'he', 'more', 'his', 'up', 'its', 'were', 'about'
    #, 'by', 'they', 'say', 'or', 'when', 'being', 'just', 'how'
    #, 'from', 'think', 'didnt', 'every', 'actually', 'are', 'an'
    #, 'ever', 'such', 'who', 'had', 'would', 'time', 'times'
    #, 'very', 'some', 'there', 'get', 'watching', 'too', 'which'
    #, 'really', 'even', 'js', 'see', 'still', '2', '1', 'watch'
   # , 'can', 'watched', 'feel', 'much', 'cause'
])

# Define a function to preprocess text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Return cleaned text
    return text

# Apply preprocessing to the 'Review' column
review_data['clean_review'] = review_data['Review'].apply(lambda x: preprocess_text(str(x)))

# Tokenization and counting, excluding stop words
def tokenize_and_count(text):
    # Split text into words
    tokens = text.split()
    # Remove stop words
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

# Join all cleaned reviews into a single string
all_reviews = ' '.join(review_data['clean_review'])

# Tokenize and count words
all_tokens = tokenize_and_count(all_reviews)
word_counts = Counter(all_tokens)

# Convert to DataFrame and sort
word_counts_df = pd.DataFrame(word_counts.items(), columns=['Word', 'Count'])
word_counts_df = word_counts_df.sort_values(by='Count', ascending=False)

# Plot the top 20 words
plt.figure(figsize=(12, 15))
top_words = word_counts_df.head(100)

plt.barh(top_words['Word'], top_words['Count'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('Word')
plt.title('Top 100 Most Common Words in Reviews')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import requests

# Your TMDb API key
api_key = '7bb3e89e3811e1b196d1ab649b627434'  # Replace with your actual API key
base_url = 'https://api.themoviedb.org/3'

def search_movie_by_title_and_year(movie_title, movie_year):
    # Search URL to find the movie by title and year
    search_url = f"{base_url}/search/movie"
    params = {
        'api_key': api_key,
        'query': movie_title,
        'year': movie_year,
        'page': 1,
        'include_adult': False
    }
    
    # Make the request to TMDb API
    response = requests.get(search_url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data['results']:
            # Print raw data for debugging
            print("Search Results Data:", data)
            # Filter results to match the title and year exactly
            for result in data['results']:
                if 'release_date' in result and result['release_date'].startswith(str(movie_year)):
                    if result['title'].lower() == movie_title.lower():
                        return result['id'], result['title'], result['release_date']
    return None, None, None

def get_titles_in_languages(movie_id, languages):
    # Get movie details using movie ID and fetch translations
    details_url = f"{base_url}/movie/{movie_id}"
    params = {
        'api_key': api_key,
        'append_to_response': 'translations'
    }
    
    # Make the request to TMDb API
    response = requests.get(details_url, params=params)
    if response.status_code == 200:
        data = response.json()
        translations = data.get('translations', {}).get('translations', [])
        
        # Debug: Print raw translations data
        print("Translations Data:", translations)
        
        # Create a dictionary for easy lookup
        language_titles = {t['iso_639_1']: t['name'] for t in translations}
        
        # Include the original title if itâ€™s not in the translations
        original_title = data.get('original_title', 'N/A')
        if 'en' not in language_titles:
            language_titles['en'] = original_title
        
        # Fetch titles in requested languages
        titles = {lang: language_titles.get(lang, 'N/A') for lang in languages}
        return titles
    return {lang: 'N/A' for lang in languages}

def get_original_title_by_title_and_year(movie_title, movie_year, languages):
    # Fetch movie ID by title and year
    movie_id, title, release_date = search_movie_by_title_and_year(movie_title, movie_year)
    if movie_id:
        print(f"Found movie: {title} ({release_date})")
        # Get titles in specified languages using the movie ID
        return get_titles_in_languages(movie_id, languages)
    return {lang: 'N/A' for lang in languages}

# Example usage
movie_title = "Soulmate"  # Replace with your movie title
movie_year = 2023  # Replace with the movie year
languages = ['ko', 'en']  # Korean (Hangul) and English

titles = get_original_title_by_title_and_year(movie_title, movie_year, languages)

for lang, title in titles.items():
    print(f"The title in language code '{lang}' is '{title}'.")
```
